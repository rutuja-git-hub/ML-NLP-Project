```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("due_diligence (1).xlsx")
```

```python
data.columns
```

```python
df0 = data[["Due Diligence Question or Request Description"]]
df = df0.rename(columns={"Due Diligence Question or Request Description":"qu"})
df = df.dropna()
df
```

```python
sentences = list(df.iloc[:, 0])
model_name= 'bert-base-nli-mean-tokens'
```

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
```

```python

model = SentenceTransformer(model_name)
```

```python
sentences_vector = model.encode(sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(
    [sentences_vector[0]],
    sentences_vector[1:]
)
```

```python
np.argmax(similarity)
```

```python
sim = pd.DataFrame({"question":df.iloc[1:,0], "similarity":similarity.ravel()})
sim
```
